```python
df = pd.read_excel(r"C:\Users\14237\Desktop\Personal Data Projects\Manufactured Housing Project\2014-2023 MHS Data.xlsx")
df
```

Drop unnecessary columns (weight, WGTADJ, foundation, secured, all J(label))

```python
df = df.drop(columns = ["WEIGHT", "WGTADJ","FOUNDATION", "SECURED", "JSTATUS", "JPRICE", "JSQFT", "JBEDROOM", "JTITLED", "JLOCATION", "JFOUNDATION", "JSECURED"])
df
```

```python
df.drop_duplicates()
df
```

```python
df.info()
```

```python
df = df.astype({"REGION": str, "SECTIONS": str, "STATUS": str, "BEDROOMS": str, "TITLED": str, "LOCATION": str})
```

Input variables using provided key

```python
df1 = df.copy()
df1["REGION"] = df1["REGION"].replace(["1", "2", "3", "4", "5"], ["Northeast", "Midwest", "South", "West", "United States"])
df1
```

```python
df1["LOCATION"] = df1["LOCATION"].replace(["1", "3"], ["Inside Manufactured Home Communities", "Outside Manufactured Home Communities"])
df1
```

```python
df1["BEDROOMS"] = df1["BEDROOMS"].replace(["1", "3"], ["2 or less", "3 or more"])
df1
```

```python
df1["TITLED"] = df1["TITLED"].replace(["1","2","3"], ["Real Estate", "Personal Property", "Not Titled"])
df1
```

```python
df1["SECTIONS"] = df1["SECTIONS"].replace(["1","2", "3"], ["Single", "Double", "3 or more sections"])
df1
```

```python
df1 = df1.astype({"SHIPMONTH": str})
df2 = df1.copy()

df1["Year"] = df1["SHIPMONTH"].str.slice(stop = 4)
df1["Month"] = df1["SHIPMONTH"].str.slice(start = 4)
df1
```

```python
df1 = df1.astype({"Year": int})
df1.loc[(df1["Year"] <= 2020) & (df1["STATUS"] == "1"), "STATUS"] = "Intended for Sale for residential use"
df1.loc[(df1["Year"] <= 2020) & (df1["STATUS"] == "2"), "STATUS"] = "Placed/Sold for residential use"
df1.loc[(df1["Year"] >=2014) & (df1["STATUS"] == "3"), "STATUS"] = "Non-residential/Other"
df1.loc[(df1["Year"] >= 2021) & (df1["STATUS"] == "1"), "STATUS"] = "Sold/Leased for residential use"
df1.loc[(df1["Year"] >= 2021) & (df1["STATUS"] == "2"), "STATUS"] = "Intended for Sale or lease for residential use"
df1
```
